```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
```

```python
# 1.Scatter Plot
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time',size='size',palette='deep')
plt.title('Scatter plot of total bill and tip')
plt.show()
```

```python
# 2.Line Plot
sns.lineplot(data=tips,x='size',y='total_bill',hue='sex',ci=None,marker='o')
plt.title('Line plot of Total Bill vs Size')
plt.show()
```

```python
tips.columns
```

```python
# 3.Bar Plot
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',palette='muted')
plt.title('Barplot of Total Bill by Day and Sex')
plt.show()
```

```python
# 4.Box Plo
sns.boxplot(data=tips,x='day',y='tip',hue='smoker',palette='Set2')
plt.title('Boxplot of tips by Day and Smoker')
plt.show()
```

```python
 # 5.violin Plot
 sns.violinplot(data=tips,x='day',y='total_bill',hue='time',split=True,palette='pastel')
 plt.title('Violin Plot of Total bill by day and time')
 plt.show()
```

```python
# 6.Count Plot
sns.countplot(data=tips,x='day',hue='smoker',palette='dark')
plt.title('Count Plot of Days by Smoker Status')
plt.show()
```

```python

tips.columns
```

```python
# 7.regression Plot
sns.regplot(data=tips,x='total_bill',y='tip',scatter_kws={'s':50},line_kws={'color':'red'})
plt.title('Regression Plot of Total Bill and Tip')
plt.show()
```

```python
 # 8.Histogram
 sns.histplot(data=tips,x='total_bill',bins=20,kde=True,color='pink')
 plt.title('Histogram of Total Bill with KDE')
 plt.show()
```

```python
 # 9.Pair Plot
 sns.pairplot(data=tips,hue='sex',vars=['total_bill','tip','size'],palette='husl')
 plt.suptitle("pair plot: numerberic variables by gender",y=1.02)    # here use a suptitle not a subtital of the graph
 plt.show()
```

```python
# 10.Cap Plot
sns.catplot(data=tips,x='day',y='tip',hue='sex',kind='point',palette='bright')
plt.title("catplot(Point):Tips by day and gender")
plt.show()
```

```python
# 11.Joint Plot
sns.jointplot(data=tips, x="total_bill", y="tip",kind='scatter',hue="sex",color='red',palette='coolwarm')
plt.title("Joint plot: Total bill vs Tip",y=1.02)
plt.show()
    
```

```python
 # 12.facetgrid 
sns.FacetGrid(data=tips,col='time',row='smoker',margin_titles=True).map(sns.histplot,'total_bill',bins=10,kde=True).add_legend()
plt.show()
```

```python
# 13.strip plot
sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter=True, palette='Set1')
plt.title("strip plot: Tips by data and gender")
plt.show()
```

```python
# 14.KDE plot
sns.kdeplot(data=tips, x='total_bill',hue='sex', fill=True, palette='tab10')
plt.title("kde plot:Total bill density by gender")
plt.show()
```
